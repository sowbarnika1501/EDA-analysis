```python
dataset["steroid"].fillna(dataset["steroid"].mode()[0],inplace=True)
dataset["fatigue"].fillna(dataset["fatigue"].mode()[0],inplace=True)
dataset["malaise"].fillna(dataset["malaise"].mode()[0],inplace=True)
dataset["anorexia"].fillna(dataset["anorexia"].mode()[0],inplace=True)
dataset["liver_big"].fillna(dataset["liver_big"].mode()[0],inplace=True)
dataset["liver_firm"].fillna(dataset["liver_firm"].mode()[0],inplace=True)
dataset["spleen_palpable"].fillna(dataset["spleen_palpable"].mode()[0],inplace=True)
dataset["spiders"].fillna(dataset["spiders"].mode()[0],inplace=True)
dataset["ascites"].fillna(dataset["ascites"].mode()[0],inplace=True)
dataset["varices"].fillna(dataset["varices"].mode()[0],inplace=True)
dataset["bilirubin"].fillna(dataset["bilirubin"].mean(),inplace=True)
dataset["alk_phosphate"].fillna(dataset["alk_phosphate"].mean(),inplace=True)
dataset["sgot"].fillna(dataset["sgot"].mean(),inplace=True)
dataset["albumin"].fillna(dataset["albumin"].mean(),inplace=True)
dataset["protime"].fillna(dataset["protime"].mean(),inplace=True)

dataset
```

```python
x= dataset.iloc[:,0:18].values
x
```

```python
x.shape
```

```python
y = dataset.iloc[:,-1].values
y
```

```python
y.shape
```

```python
from sklearn.model_selection import train_test_split
x_train,x_test,y_train,y_test=train_test_split(x,y,test_size=0.2,random_state=0)
```

```python
x_train.shape
```

```python
y_train.shape
```

```python
x_test.shape
```

```python
y_test.shape
```

```python
from sklearn.ensemble import RandomForestClassifier
rrc=RandomForestClassifier(n_estimators=100,criterion='entropy',random_state=0)
rrc.fit(x_train,y_train)
```

```python
y_pred=rrc.predict(x_test)
y_pred
```

```python
from sklearn.metrics import accuracy_score
accuracy_score(y_test,y_pred)
```

```python
from sklearn.metrics import confusion_matrix
cm = confusion_matrix(y_test,y_pred)
cm
```

```python
from sklearn.metrics import recall_score, precision_score
print("Precision score : %f" %(precision_score(y_test, y_pred) * 100))
```
